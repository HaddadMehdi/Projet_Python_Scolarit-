```python
import pymssql
from tkinter import *
import tkinter.font as tkFont
from tkinter.ttk import *
from tkinter.messagebox import *
import sys

import traceback

#************************** parametre de connexion à la base sql server *****************
#****************************************************************************************

SERVER = "PC-HADDAD"
USER = "sa"
PASSWORD = "sa123"
DATABASE = "Etudiants"

connection = pymssql.connect(server=SERVER, user=USER, 
                  password=PASSWORD, database=DATABASE)
 
# *************************** Parametre du fenetre principale ****************************
#****************************************************************************************
fenetre = Tk()
fenetre.title('Gestion de Scolarité')
fenetre.geometry("1047x714")
fenetre.iconbitmap('Logo-educ.ico')
fenetre.minsize(1047,714)
fenetre.config(background='#FAFAFA')


Label(fenetre, text="Gestion de Scolarité",background='#FAFAFA',font='Helvetica').pack()

#************************** Ajouter image en arriere Plan du Menu Principale****************
#****************************************************************************************
photo = PhotoImage(file="Education.png")
canvas = Canvas(fenetre,width=708, height=368)
canvas.create_image(0, 0, anchor=NW, image=photo)
canvas.pack(fill='both',expand=1 )




#**************************** affichage du Fenetre Crerer Nouveau Etudiants ************************************************
#************************************************************************************************************ 
    
def create():
    import tkinter as tk
    import pymssql
    
    import tkinter.font as tkFont
    

    import sys

    import traceback 
    
    fenetre2 = Tk()
    fenetre2.title('Ajouter un Etudiant') 
    
    # libellé
    libelle = Label(fenetre2, text='Nom : ')
    libelle.grid(row=0, column=0, sticky=E, padx=10, pady=20)     
    name = Entry(fenetre2,width=30)
    name.focus_set()
    name.grid(row=0, column=1, padx=(0, 15), pady=10)
    
    
    
    libelle = Label(fenetre2, text='Prénom : ')
    libelle.grid(row=0, column=2, sticky=E, padx=10, pady=10)   
    Prenom = Entry(fenetre2,width=30)
    Prenom.focus_set()
    Prenom.grid(row=0, column=3, padx=(0, 15), pady=10)
    
    libelle = Label(fenetre2, text='Adresse : ')
    libelle.grid(row=2, column=0, sticky=E, padx=10, pady=10)     
    Adresse = Entry(fenetre2,width=30)
    Adresse.focus_set()
    Adresse.grid(row=2, column=1, padx=(0, 15), pady=10)
    
    libelle = Label(fenetre2, text='Ville : ')
    libelle.grid(row=2, column=2, sticky=E, padx=10, pady=10)     
    Ville = Entry(fenetre2,width=30)
    Ville.focus_set()
    Ville.grid(row=2, column=3, padx=(0, 15), pady=10)
    
    libelle = Label(fenetre2, text='Téléphone Fixe : ')
    libelle.grid(row=3, column=0, sticky=E, padx=10, pady=10)     
    TelFixe = Entry(fenetre2,width=30)
    TelFixe.focus_set()
    TelFixe.grid(row=3, column=1, padx=(0, 15), pady=10)
    
    libelle = Label(fenetre2, text='Téléphone Mobile : ')
    libelle.grid(row=3, column=2, sticky=E, padx=10, pady=10)     
    TelMobile = Entry(fenetre2,width=30)
    TelMobile.focus_set()
    TelMobile.grid(row=3, column=3, padx=(0, 15), pady=10)
    
    
    
    
    #**************************** Fonction pour inserer nouveau etudiant dans la base de données ************************************************
    #********************************************************************************************************************************************
    
    def getEntry():
        
        import pymssql
        from tkinter import messagebox
        
        res1 = name.get()
        res2 = Prenom.get()
        res3 = Adresse.get()
        res4 = Ville.get()
        res5 = TelFixe.get()
        res6 = TelMobile.get()
        
        SERVER = "PC-HADDAD"
        USER = "sa"
        PASSWORD = "sa123"
        DATABASE = "Etudiants"
        connection = pymssql.connect(server=SERVER, user=USER,password=PASSWORD, database=DATABASE)
        cursor = connection.cursor()
        cursor2 = connection.cursor()
        sql2="SELECT MAX([NumeroEtudiant]) FROM [Etudiants].[dbo].[etudiants]"
        cursor2.execute(sql2)
        row2 = cursor2.fetchone()
        DernierID=row2[0] 
        
        sql ="INSERT INTO [Etudiants].[dbo].[etudiants] ([NumeroEtudiant],[Prenom],[Nom],[Adresse],[Ville],[TelephoneFixe],[TelephonePortable]) VALUES (%s,%s,%s,%s,%s,%s,%s)"
        cursor.execute(sql,(DernierID+1,res1,res2,res3,res4,res5,res6))
        connection.commit()
        connection.close()
        name.delete(0,END)
        Prenom.delete(0,END)
        Adresse.delete(0,END)
        Ville.delete(0,END)
        TelFixe.delete(0,END)
        TelMobile.delete(0,END)
        messagebox.showinfo("Message info", "Integraion Réussie")
        
    
           
    button = Button( fenetre2, text="Integrer",command=getEntry)
    button.grid(row=4, column=4, padx=(0, 15), pady=10)
    
        
    fenetre2.mainloop()
    
#**************************** affichage du Fenetre Etudiants ************************************************
#************************************************************************************************************  

def Etudiant():
    import tkinter as tk
    fenetre3 = Tk()
    fenetre3.title('Etudiants') 
    fenetre3.geometry("900x520")
    
    SERVER = "PC-HADDAD"
    USER = "sa"
    PASSWORD = "sa123"
    DATABASE = "Etudiants"

    connection = pymssql.connect(server=SERVER, user=USER, 
                  password=PASSWORD, database=DATABASE)

    cursor = connection.cursor()

    cursor.execute("SELECT [NumeroEtudiant],[Prenom],[Nom],[Adresse],[Ville],[TelephoneFixe],[TelephonePortable] FROM [Etudiants].[dbo].[etudiants] ORDER BY [NumeroEtudiant] DESC ")
    row = cursor.fetchall()
    
      
    
              
    
    
    tableau = Treeview(fenetre3, columns=('NumeroEtudiant','Prenom','Nom','Adresse','Ville','TelephoneFixe','TelephonePortable'))
    
    tableau.heading('NumeroEtudiant', text='Numero Etudiant')
    tableau.heading('Prenom', text='Prenom')
    tableau.heading('Nom', text='Nom')
    tableau.heading('Adresse', text='Adresse')
    tableau.heading('Ville', text='Ville')
    tableau.heading('TelephoneFixe', text='Telephone Fixe')
    tableau.heading('TelephonePortable', text='Telephone Portable')
    
    Label(fenetre3, text="Liste des Etudiant",background='#FAFAFA',font='Helvetica',foreground='#006064').pack(fill='both',padx = 0,pady =(0, 0))
    
    tableau['show'] = 'headings' 
    
    
    tableau.pack(fill='both',expand=1 ,padx = 10,pady = (20, 10))
    
    
    if len(row):   
        for enreg in row:
            tableau.insert('', 'end', iid=enreg[0], values=(enreg[0],enreg[1],enreg[2],enreg[3],enreg[4],enreg[5],enreg[6]))
   
    
    
    fenetre3.mainloop()
    
#**************************** affichage du Fenetre Résultats ************************************************
#************************************************************************************************************
def Resultat():
    import tkinter as tk
        
    fenetre4 = Tk()
    fenetre4.title('Resultats') 
    
    # libellé
    Label(fenetre4, text="Imprimer résultat Etudiant",background='#FAFAFA',font='Helvetica',foreground='#006064').grid(row=0, column=0, sticky=E, padx=10, pady=20)
    
    libelle = Label(fenetre4, text='Numero Etudiant : ')
    libelle.grid(row=1, column=0, sticky=E, padx=10, pady=20)     
    code = Entry(fenetre4,width=30)
    code.focus_set()
    code.grid(row=1, column=1, padx=(0, 15), pady=10)
      
    #********************************** Fonction pour extraire les données de la table résultat****************************
    #************************************************************************************************************************
    
    def get_resultat():
        import pymssql
        import tkinter.font as tkFont
        import sys
        import traceback
        import win32api 
        from tkinter import filedialog
        
        SERVER = "PC-HADDAD"
        USER = "sa"
        PASSWORD = "sa123"
        DATABASE = "Etudiants"

        connection4 = pymssql.connect(server=SERVER, user=USER, 
                      password=PASSWORD, database=DATABASE)
        connection5 = pymssql.connect(server=SERVER, user=USER, 
                      password=PASSWORD, database=DATABASE)
        cursor4 = connection4.cursor()
        cursor5 = connection5.cursor()
        CodeE = code.get()
        sql4="SELECT  [Etudiants].[dbo].[inscriptions].[NumeroInscription], AVG([Etudiants].[dbo].[resultats].[Note]) AS Expr1, [Etudiants].[dbo].[inscriptions].[NumeroEtudiant], [Etudiants].[dbo].[matieres].[CodeMatiere] FROM   [Etudiants].[dbo].[inscriptions] INNER JOIN [Etudiants].[dbo].[resultats] ON [Etudiants].[dbo].[inscriptions].[NumeroInscription] = [Etudiants].[dbo].[resultats].[NumeroInscription] INNER JOIN [Etudiants].[dbo].[matieres] ON [Etudiants].[dbo].[inscriptions].[CodeMatiere] = [Etudiants].[dbo].[matieres].[CodeMatiere] WHERE  [Etudiants].[dbo].[inscriptions].[NumeroEtudiant] ="+str(CodeE)+" GROUP BY [Etudiants].[dbo].[inscriptions].[NumeroInscription], [Etudiants].[dbo].[inscriptions].[NumeroEtudiant], [Etudiants].[dbo].[matieres].[CodeMatiere]"
        sql5="SELECT  [NumeroEtudiant],[Prenom],[Nom],[Adresse],[Ville],[TelephoneFixe],[TelephonePortable]  FROM [Etudiants].[dbo].[etudiants] WHERE [NumeroEtudiant] ="+str(CodeE)
        cursor4.execute(sql4)
        cursor5.execute(sql5)
        row4 = cursor4.fetchall()
        row5 = cursor5.fetchall()
        Valeur4=row4
        
        
        #********************************* Save les résultat dans un Fichier texte nommé bulletin et l'imprimer ************************
        #****************************************************************************************************************
        def save_to_file(text):
            with open('Bulletin Etudiant N° '+CodeE+'.txt', mode='w') as myfile:
                if len(row4):
                    
                    myfile.write('\n')
                    myfile.write('Université Claude Chappe'+'\n')
                    myfile.write('15 avenue de Moulincourbe'+'\n')
                    myfile.write('28094 Clairecombe'+'\n')
                    for enreg2 in row5:
                        myfile.write('                                                   '+str(enreg2[1])+' '+str(enreg2[2])+'\n')
                        myfile.write('                                                   '+str(enreg2[3])+' '+str(enreg2[4]))
                        myfile.write('\n')
                                               
                    myfile.write('\n')
                    myfile.write('\n')
                    myfile.write('\n')
                    myfile.write('Madame, Monsieur'+'\n') 
                    myfile.write('Veuillez trouver dans le récapitulatif ci-dessous les résultats de vos examens.'+'\n') 
                    myfile.write('\n')
                    myfile.write('Matière'+'           '+'Moyenne')
                    myfile.write('\n')
                    
                    for enreg in row4:
                        myfile.write(str(enreg[3])+'        '+str(enreg[1]))
                        myfile.write('\n')
                    myfile.write('\n')
                    myfile.write('\n')
                    myfile.write('Ce document constitue les résultats officiels. Pour toute contestation, contactez le service scolarité.')
        save_to_file(Valeur4)
        
        #********************************** Imprimer Bulletin de l'étudiant  *********************************
        #*****************************************************************************************************
        file_to_print = filedialog.askopenfilename(initialdir="C:/Users/MED MEHDI HADDAD",initialfile="Bulletin Etudiant N° "+CodeE+".txt", title="Select file",filetypes=(("Text files", "*.txt"), ("all files", "*.*")))   
        if file_to_print: 
            win32api.ShellExecute(0, "print", file_to_print, None, ".", 0)  
  

        
    button = Button( fenetre4, text="Imprimer",command=get_resultat)
    button.grid(row=2, column=1, padx=(0, 15), pady=10)
    
    fenetre4.mainloop()
    
    
# ********************** Création du Menu de la Fenetre Principale ********************
#****************************************************************************************

menubar = Menu(fenetre)
menubar.config(background='#D1C4E9')

photo1 = PhotoImage(file = "Education 1.png")  
photoimage = photo1.subsample(3, 3) 
photoE = PhotoImage(file = "Etudiant.png")  
photoimageE = photoE.subsample(3, 3) 
photoExit = PhotoImage(file = "Exit.png")  
photoimageExit = photoExit.subsample(3, 3)  


menu1 = Menu(menubar, tearoff=0)
menu1.add_command(label="Etudiants",image = photoimageE,compound = LEFT, command=Etudiant)
menu1.add_command(label="Ajouter Etudiant",image = photoimage,compound = LEFT,command=create)
menu1.add_command(label="Resultats",command=Resultat)
menu1.add_separator()
menu1.add_command(label="Quitter",image = photoimageExit,compound = LEFT, command = fenetre.destroy)
menubar.add_cascade(label="Fichier", menu=menu1)
menu1.config(background='#D1C4E9')

fenetre.config(menu=menubar)

            
# ************************** Deconnexion de la base sql***********************            
if connection:

    connection.close()
    
fenetre.mainloop()


```
